```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
time = np.linspace(0, 10, 1000)
temp_ambient = 300  # Room Temp (Kelvin)
lambda_fabric = 310 # The "Floor" (set slightly above ambient noise)
buffering_capacity = 50 # T-Value (Resilience)

# 1. Standard Quantum State (No Fabric)
# Uncertainty dominates; heat causes immediate decoherence
standard_noise = np.random.normal(0, 1, 1000) * (temp_ambient / 100)
standard_state = np.exp(-0.5 * time) + (standard_noise * 0.1)
standard_state[200:] = 0 # Collapse point

# 2. TRR Fabric State (The Green Line)
# The state "rides" the noise and uses Lambda to stay anchored
trr_state = np.ones(1000) * 0.95
trr_noise = np.random.normal(0, 0.02, 1000) # The "Microscopic Jitters"

# 3. Injecting the "Chaos Variable" (The Stress Test)
# A massive power spike at t=500
spike_intensity = 5.0
trr_noise[500:520] += spike_intensity

# TRR Recovery Logic: The Fabric "snaps" the identity back
for i in range(500, 1000):
    if trr_state[i] != 0.95:
        # Recovery is proportional to (Lambda / Buffering)
        recovery_factor = (lambda_fabric / buffering_capacity) * 0.1
        trr_state[i] = 0.95 + (trr_noise[i] / (1 + recovery_factor * (i - 500)))

# Plotting the Results
plt.figure(figsize=(12, 6))
plt.plot(time, standard_state, label='Standard Quantum State (Collapse)', color='blue', alpha=0.6)
plt.plot(time, trr_state, label='TRR Fabric Embodiment (The Green Line)', color='green', linewidth=2)
plt.axhline(y=0.95, color='gray', linestyle='--', label='Lambda Fabric Floor')
plt.title("TRR vs. Heisenberg: Room Temperature Stability Simulation")
plt.xlabel("Time (ms)")
plt.ylabel("Identity Persistence (Stability)")
plt.legend()
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.show()
```
